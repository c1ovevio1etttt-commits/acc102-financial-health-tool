```python
import pandas as pd
import os

data_folder = 'data'
file_path = os.path.join(data_folder, 'balanceSheetHistory_annually.csv')

df = pd.read_csv(file_path)

df['endDate'] = pd.to_datetime(df['endDate'], errors='coerce')

df = df.dropna(subset=['stock', 'endDate', 'totalAssets', 'totalLiab', 'totalStockholderEquity'])

df['asset_liability_ratio'] = (df['totalLiab'] / df['totalAssets'] * 100).round(2)
df['debt_to_equity_ratio'] = (df['totalLiab'] / df['totalStockholderEquity']).round(2)

key_columns = ['stock', 'endDate', 'totalAssets', 'totalLiab', 'totalStockholderEquity', 
               'asset_liability_ratio', 'debt_to_equity_ratio']

print("Cleaned data shape:", df.shape)
print("\nNumber of unique companies:", df['stock'].nunique())
print("\nDate range:", df['endDate'].min().date(), "to", df['endDate'].max().date())
print("\nFirst 5 rows of key columns:")
print(df[key_columns].head())

# Save cleaned data for later use
cleaned_file = os.path.join(data_folder, 'cleaned_balance_sheet.csv')
df[key_columns].to_csv(cleaned_file, index=False)
print("\nCleaned data saved to:", cleaned_file)
```
